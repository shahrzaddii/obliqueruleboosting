```python
import numpy as np
from sklearn.metrics import mean_squared_error, log_loss, zero_one_loss
import pandas as pd
import lltboost
from dataset import CallData
```

```python
dataset = CallData().call(dataset_name = 'diabetes', data_format = 'df')
X = dataset.X
y = dataset.y
```

```python
lltb_model = lltboost.LinRepRuleEnsemble(num_rules=6, task=dataset.learning_type, max_complexity=5, reg=1e-4).fit(X,y)
lltb_model.rules(), lltb_model.model_complexity(),  mean_squared_error(y, lltb_model.predict(X))
```
